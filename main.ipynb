##Converting text files to csv format


```python
import os
import csv
directory_path = "Radar_Point_Cloud_Data"
filenames = os.listdir(directory_path)
print(filenames)
for file in filenames:
    fileopen = open(os.path.join(directory_path,file), 'r')
    file_new = os.path.join(os.path.join(directory_path,file))
    file_new = file_new.split('.')[0]
    file_new = file_new+'.csv'
    lines = fileopen.read().splitlines()
    for line in lines:
        data = line.split()
        try:
            with open(file_new, 'a', newline='') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow(data)
        except FileNotFoundError:
            with open(file_new, 'w', newline='') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow(data)
```

RADAR DATA consists of:
        
        
        X                    # X m
        Y                    # Y m
        Z                    # Z m
        v                    # Doppler Velocity m/s
        Rindex               # Point range index in Range_Doppler Matrix
        Dindex               # Point Doppler index in Range_Doppler Matrix
        Peakval              # Peak Value
        pitchangle           # Angle from radar to target point
        projectX             # Pixel index X when project the points to the image.
        projectY             # Pixel index Y when project the points to the image.
        annotation           # Annotations for the point cloud data. (0 means the point is of real target while 1 means not)

Getting (x,y,z,v,p) from the above data

```python
import os
import csv
directory_path = "Radar_Point_Cloud_Data"
filenames = os.listdir(directory_path)
print(filenames)
for file in filenames:
    fileopen = open(os.path.join(directory_path,file), 'r')
    file_new = os.path.join(os.path.join(directory_path,file))
    file_new = file_new.split('.')[0]
    file_new = file_new+'.csv'
    lines = fileopen.read().splitlines()
    for line in lines:
        data = line.split()
        selected_elements = data[0:4]
        selected_elements.append(data[6])

        try:
            with open(file_new, 'a', newline='') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow(selected_elements)
        except FileNotFoundError:
            with open(file_new, 'w', newline='') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow(selected_elements)
```
